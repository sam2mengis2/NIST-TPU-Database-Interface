```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
test_numbers = np.array([1,2,3,4,5,6,7,8]) #test number for the phase 1

roof_slope = np.array([1,3,0.25,1,1,0.25,0.25,0.25]) #out of 12 
plt.scatter(test_numbers, roof_slope, color='blue', label='Roof Slope (x/12)')

# 2. Add labels so the graph makes sense for your WindEEE report
plt.xlabel('Test Number')
plt.ylabel('Roof Slope (out of 12)')
plt.title('NIST/TPU Dataset Slope Distribution')
plt.grid(True)

# 3. Finally, show the result
plt.show()
```
