```python
import subprocess as sb
import os
```

```python
decoder_path = r"C:\WINDLAB_SUMMER\NIST_Download\bin\win32\hdf_read_demo_win.exe"
file_path = r"C:\WINDLAB_SUMMER\file_drop\ADW100o100D016a1800.HDF"

print(f"Decoder exists: {os.path.exists(decoder_path)}")
print(f"HDF file exists: {os.path.exists(file_path)}")
```

```python
result = sb.run(
    [decoder_path, file_path],
    capture_output=True,
    text=True
)

print("--- STDOUT ---")
print(result.stdout)
print("--- STDERR ---")
print(result.stderr)
print(f"Return Code: {result.returncode}")
```
